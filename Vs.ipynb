```python
# Install necessary libraries
!pip install flask flask-ngrok pydub ffmpeg-python
!apt-get update
!apt-get install ffmpeg
!pip install git+https://github.com/openai/whisper.git
!pip install deep-translator

from flask import Flask, request, jsonify
import whisper
import ffmpeg
from deep_translator import GoogleTranslator
import os
import requests
import time
from threading import Thread

# Initialize Flask app
app = Flask(__name__)

# Initialize Whisper model
model = whisper.load_model("base")

@app.route('/')
def home():
    return "Server is Running!"

@app.route('/process', methods=['POST'])
def process_video():
    if 'video' not in request.files:
        return jsonify({"error": "No video uploaded"}), 400

    # Save the video file
    file = request.files['video']
    video_path = "/content/input_video.mp4"
    audio_path = "/content/audio.wav"
    file.save(video_path)

    # Extract audio from video using FFmpeg
    ffmpeg.input(video_path).output(audio_path).run(overwrite_output=True)

    # Transcribe audio to text using Whisper
    result = model.transcribe(audio_path)
    text = result["text"]

    # Translate text to Bengali using Deep Translator
    translated_text = GoogleTranslator(source='auto', target='bn').translate(text)

    return jsonify({"translated_text": translated_text})

# Function to run flask
def run_flask():
    app.run(port=5000)

# Install ngrok and set up the token
!wget -q -nc https://bin.equinox.io/c/bNyj1mQVY4c/ngrok-stable-linux-amd64.zip
!unzip -o ngrok-stable-linux-amd64.zip
!mv -f ngrok /usr/local/bin

# Set ngrok auth token (only once needed)
!ngrok config add-authtoken 2wAdb14fQ9S9aXgcAfb9eNf9fVG_5KRft9aW4XpseC4YcFzsE

# Start flask server in background
Thread(target=run_flask).start()

# Wait a bit
time.sleep(5)

# Start ngrok
os.system("ngrok http 5000 &")
time.sleep(4)

# Get ngrok public url
r = requests.get('http://localhost:4040/api/tunnels')
public_url = r.json()['tunnels'][0]['public_url']
print("Public URL:", public_url)
```
